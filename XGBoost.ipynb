```python
import pandas as pd
df = pd.read_csv('heart_disease_uci.csv')
df.head(40)
```

```python
df.isna().sum()
df.info()
df.columns
```

```python
df["target"] = (df["num"] > 0).astype(int)
df = df.drop('num', axis=1)
df.head(10)
```

```python
# handling the categorical values
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df['thal'] = encoder.fit_transform(df[['thal']])
df['slope'] = encoder.fit_transform(df[['slope']])
df['exang'] = encoder.fit_transform(df[['exang']])
df['restecg'] = encoder.fit_transform(df[['restecg']])
df['dataset'] = encoder.fit_transform(df[['dataset']])
df['cp'] = encoder.fit_transform(df[['cp']])
df['trestbps'] = encoder.fit_transform(df[['trestbps']])
df['sex'] = encoder.fit_transform(df[['sex']])
```

```python
# handling missing values using simpleImputer
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df[['trestbps']] = imputer.fit_transform(df[['trestbps']])
df[['chol']] = imputer.fit_transform(df[['chol']])
df[['fbs']] = imputer.fit_transform(df[['fbs']])
df[['thalch']] = imputer.fit_transform(df[['thalch']])
df[['oldpeak']] = imputer.fit_transform(df[['oldpeak']])

df[['ca']] = imputer.fit_transform(df[['ca']])
df[['slope']] = imputer.fit_transform(df[['slope']])
df[['thal']] = imputer.fit_transform(df[['thal']])
df[['exang']] = imputer.fit_transform(df[['exang']])
df[['restecg']] = imputer.fit_transform(df[['restecg']])

df.isna().sum()
```

```python
for col in ["trestbps", "chol"]:
    median_value = df[col].median()
    df[col] = df[col].replace(0, median_value)
    
df = df.drop(columns=["slope", "ca", "thal", "dataset", "id"])
```

```python
import seaborn as sns
sns.heatmap(df.corr(), cmap='viridis')
df.corr()
```

```python
# feature enginnering
df['exang*2'] = df['exang'] ** 2
# df['oldpeak*2'] = df['oldpeak'] ** 2
df.head(20)
```

```python
# independent and dependent features 
x = df.drop('target', axis=1)
y = df['target']
# spliting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# model selection 
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
xgbc = XGBClassifier()
randomf = RandomForestClassifier()
```

```python
parameters = {
    "n_estimators": [100, 200, 300],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1, 0.2],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    "gamma": [0.1, 0.5, 1.0]
}

parameters_F = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 20, 30, None],
    'max_features': ['sqrt', 'log2', 0.5]
}
```

```python
# hyper paramater tuning
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=xgbc, param_grid=parameters, scoring='f1', cv=5,  n_jobs=-1, verbose=True)
grid.fit(x_train, y_train)
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
print(f"confusion matrix: {confusion_matrix(y_test, y_pred)} \n")
print(f"accuracy: {accuracy_score(y_test, y_pred)} \n")
print(f"f1_score: {f1_score(y_test, y_pred)} \n")
print(f"report: {classification_report(y_test, y_pred)}")
```
